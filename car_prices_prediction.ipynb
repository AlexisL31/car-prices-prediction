Introduction

This notebook will focus on analysing and preparing the data of real world car prices dataset with the aim of transforming the raw data into a clean and structured form. The dataset contains information about the different types of cars and their prices alongside a mixture of categorical attributes  describing their specifications, characteristics and market factors. 

The primary objective of this analysis is to critically assess the structure and quality of the data and apply appropriate processing techniques that includes identifying missing values or duplicated records, correcting data types and transforming variables where necessary. 


```python
#Loading the Dataset

import pandas as pd        #Pandas are used for handling data structures like tables and performs data cleaning tasks
import numpy as np         #NumPY provides numerical operations thjat are required when working with real world datasets 

df= pd.read_csv("CarPrices2026.csv")  # This loads the dataset into a Pandas dataframe to enable analysis.
df.head()                             # Displaying the first few rows helps verify that the file has read correctly and the values appear in expected formats

```

```python
# Data Inspection

df.info()         # This provides an overview of the dataset by examining the column names, data types and non missing value counts. 



# Currently it shows that the dataset has 8,130 rows and 12 columns. 
# Columns that contain no missing values are Name, Year, Selling Price, Km driven, Fuel, Seller type, Transmission and Owner.
# Columns with missing values are Mileage, Engine, Max power and seats.
```

```python
# Descriptive Analysis

df.describe()      # Using this provides an overiew of the dataset, including measuress such as the mean, standard deviation, minimum and maximum values
                   # By doing so it allows us to identify and detect anomalies, data entry errors and outliers. 

```

```python
# Missing Values Check

df.isnull().sum()   # Identifies the missing values 

```

```python
# Duplicate Value Check

df.duplicated().sum()
```

```python
# Unique Value Analysis

# This function evaluates the variability of each feature by showing the number of distinct values as a percentage of total observations.
# Helping identify features that provide limited information due to low variability. 

unique_counts = df.nunique()
unique_percentage = (unique_counts / len(df)) *100
unique_percentage.sort_values()
```

The results show us that some columns have very few different values but in this dataset's case it is expected. Features like transmission, fuel type, seller type and ownership fall into a small number of categories but still provide useful information about each car. Having these values does not mean that the data is poor or unhelpful but it reflects how cars are classified in the real world. Therefore there is no need to remove any of these features and the columns are kept for further analysis.

Addressing the Missing Values

During the missing values check most of the dataset is complete, but four of the columns have missing information. The mileage, engine size, max power and number of seats each have 221 missing values while all the other columns have full data. Since these features are important aspects of a car, removing the rows with missing values will affect the analysis because the information that will be removed are useful features of a car. Instead, these missing values need to be analysed carefully so that the dataset stays complete and useful for further analysis.


```python
# Handling the Missing Values


df["max_power"] = df["max_power"].str.replace("bhp","",) # It is essential that the value of Max power is changed from an object to float as it will not allow calculations to be performed correctly when the data is stored as text rather than a numerical value.  
df["max_power"] = df["max_power"].str.strip() # Removes any leading or trailing spaces
df["max_power"] = df["max_power"].replace("",np.nan) # replaces empty strings with NaN
df["max_power"] = df["max_power"].astype(float) # Applying this code ensures that the value is turned from an object to a float to allow calculations.              


# Missing values were handled by using median as the calculation because it provides a central value that is less affected by outliers in the data.

df["max_power"] = df["max_power"].fillna(df["max_power"].median())
df["mileage(km/ltr/kg)"] = df["mileage(km/ltr/kg)"].fillna(df["mileage(km/ltr/kg)"].median())
df["engine"] = df["engine"].fillna(df["engine"].median())
df["seats"] = df["seats"].fillna(df["seats"].median())


df.isnull().sum() # This tells us now that there are no longer missing values and that all the columns are turned into the correct format. 
               
```

```python
df.info() # This final check shows that there are no incomplete records, missing values have successfully been addressed and that the features are stored correctly in its appropriate data type. . 
```

```python
# Defining Target and Features

y = df["selling_price"] # Target variable

x = df.drop(["selling_price","name"], axis=1) # dropping columns that are not useful

print(x.shape)
print(y.shape)
```

```python
# Encoding categorical variables

x = pd.get_dummies(x, drop_first = True) # Converts text columns to numeric columns for machine learning

print(x.shape)
```

```python
# Train test split

from sklearn.model_selection import train_test_split

# Dataset is split into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x,y, test_size=0.2, random_state = 99
)

print(x_train.shape, x_test.shape) # checks the shape of the data
```

```python
# linear Regression

from sklearn.linear_model import LinearRegression

lr = LinearRegression() # Creates the model
lr.fit(x_train , y_train) # trains the model using the data

y_pred_lr = lr.predict(x_test) # predicts on test data
```

```python
# Evaulating Lineare Regression

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred_lr) # calculates the mean squared error
rmse = np.sqrt(mse) # converts MSE to RMSE by square root

print("Linear Regression results:")
print("MAE:", mean_absolute_error (y_test, y_pred_lr)) # Average error
print("RMSE:", rmse) # Penalises large errors
print("R2:", r2_score(y_test, y_pred_lr)) # model performance
```

```python
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state = 99) # Creates the model
rf.fit(x_train, y_train) # Trains the model

y_pred_rf = rf.predict(x_test) # Predicts the test data
```

```python
# Evaluates the random forest

mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt (mse_rf)

print("Random Forest results:")
print("MAEl", mean_squared_error(y_test, y_pred_rf))
print("RMSE:", rmse_rf)
print("R2", r2_score(y_test, y_pred_rf))
```

```python
# Gradient boosting

from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(random_state = 99) # Creates the Gradient boosting model
gb.fit(x_train, y_train) # Trains the model

y_pred_gb = gb.predict(x_test) # Predicts the test data
```

```python
# Evaluates the Gradient Boosting

mse_gb = mean_squared_error(y_test, y_pred_gb)
rmse_gb = np.sqrt(mse_gb)

print("Gradient Boosting Results:")
print("MAE:", mean_absolute_error(y_test, y_pred_gb))
print("RMSE:", rmse_gb)
print("R2:", r2_score(y_test, y_pred_gb))
```

```python
# Comparison Table


# Creates comparison table of all models
results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "Gradient Boosting"],
    "R2": [
        r2_score(y_test, y_pred_lr),
        r2_score(y_test, y_pred_rf),
        r2_score(y_test, y_pred_gb)
    ],
    "RMSE": [
        rmse,
        rmse_rf,
        rmse_gb
    ]
})

print(results)
```
